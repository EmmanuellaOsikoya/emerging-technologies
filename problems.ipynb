## Problem 1: Generating Random Boolean Functions

### In order for me to be able to generate random boolean functions, I am going to use the Deutsch-Jozsa algorithm. The Deutsch-Jozsa algorithm is an algorithm that works with special kinds of functions. Each function takes serveral Boolean values as input and then returns either true or false. The function is guaranteed to be either a constant function which always returns the same value, or a balanced function. The output will depend on whether the function chosen was constant or balanced.

### Before creating constant or balanced functions, we need all possible inputs. Each input can be etiher true or false. In this case we have four inputs (a,b,c,d). So the total number of combinations would be 16 (2^4). Examples of the possible combinations would be 1. true, true, true, true, 2. false, false, false, false, 3. false, false, false, true and 4. false, false, true, false

```python
from itertools import product
```

```python
def generate_all_inputs():
    """
    This function is responsible for generating all possible Boolean input combinations for four inputs
    
    Each input can be either true or false we have four inputs total and we want every possible combination which would give us 16 total combinations
    
    Returns:
    
    list of tuples
    
    Each tuple contains four Boolean values
    """
    
    # I am using Python's itertools.product to create combinations
    # product([False, True], repeat=4) means that you take a false and a true repeat them 4 times and then make every possible grouping, automatically creating all 16 combinations
    
    all_combinations = list(product([False, True], repeat=4))
    
    # Then I convert it into a list so it is easier to use later
    
    return all_combinations
    
```

```python
# This is a test to see that the generate_all_inputs() function works

if __name__ == "__main__":
    inputs = generate_all_inputs()
    
    print("Total combinations:", len(inputs))
    
    for combo in inputs:
        print(combo)
```

### A constant function ignores its inputs, so no matter what values you give it it always returns true or false so the next thing that we need to do is write a function that randomly chooses True or False and then return a function that always gives that value

### The inner function remembers the chosen value. Even after the outer function finishes, the inner function still knows what value to return. This is how we keep the function constant

```python
import random

def generate_constant_function():
    """
    This function creates and returns a constant Boolean function.
    
    A constant function always returns the same value.
    It does NOT care about the input values.
    
    For example:
    
        f(True, True, True, True) -> False
        f(False, False, True, False) -> False
        
    This function will:
    
    1. Randomly choose either True or False
    2. Create a new function that always returns that value
    3. Return the new function
    
    Returns:
    
    A function that takes four Boolean inputs and always returns the same Boolean value
    """
    
    # Randomly choose True or False
    constant_value = random.choice([True, False])
    
    # Now I am going to create a function inside this function
    # This inner function will "remember" the constant_value
    # This is called a closure
    # The inner function keeps access to constant_value even after generate_constant_function finishes
    
    def constant_function(a,b,c,d):
        """
        A constant Boolean function.
        
        Parameters:
        
        a,b,c,d : bool
            Four Boolean input values
            
        Returns:
        
        bool 
        
        Always returns the same value chosen earlier
        """
        
        return constant_value
    
    # Returns the constant function we just created
    return constant_function
```

```python
# Simple tesr to check that the function is working

if __name__ == "__main__":
    
    f = generate_constant_function()
    
    print("Testing constant function:")
    
    print(f(True, True, True, True))
    print(f(False, False, False, False))
    print(f(True, False, True, False))
```

### The next thing I need to do is to create a balanced function generator. A balanced function returns true for exactly half of all inputs and false for the other half. In this problem we have a total of 16 inputs so we must return exactly 8 true and 8 false.

### The function will work as follows:

### The function will generate all 16 input Boolean combinations, shuffle them randomly, pick the first 8 to return True and the remaining 8 inputs to return false.

### The 8 True inputs are stored in a set, making checking fast and simple so if the input is in the set you return True otherwise you return False

```python
def generate_balanced_function():
    """ 
    This function is responsible for creating and returning a balanced function
    
    A balanced function returns True for exactly half of all inputs and False for the other half
    
    With four Boolean inputs there are 16 possible input combinations so a balanced function must return 8 True values and 8 False values
    
    This function will:
    
    1. Generate all 16 possible inputs
    2. Randomly choose 8 of them to return True
    3. Assign False to the remaining 8
    4. Return a function that uses this mapping
    
    Returns:
    
    function
    
    A function that takes four Bollean inputs and returns a balanced True/False result
    """
    
    # Get all possible Boolean input combinations
    all_inputs = generate_all_inputs()
    
    # Randomly shuffle the input list so the selection is random
    random.shuffle(all_inputs)
    
    # Take the first 8 inputs as True
    true_inputs = set(all_inputs[:8]) # The remaining 8 will automatically be false
    
    # Create balanced function
    def balanced_function(a,b,c,d):
        """ 
        This function is a Balanced Boolean function
        
        Parameters:
        
        a,b,c,d : bool
            Four Boolean input values
            
        Returns:
        
        bool
        
        Returns True for exactly half of all possible inputs and False for the rest
        """
        
        # Combine the inputs into a tuple
        input_tuple = (a,b,c,d)
        
        # Check if this inputs is in the True set
        if input_tuple in true_inputs:
            return True
        else:
            return False
        
    # Return the balanced function
    return balanced_function
    
```

```python
# Test to check that function is working
if __name__ == "__main__":

    f = generate_balanced_function()

    print("Testing balanced function:")

    inputs = generate_all_inputs()

    true_count = 0
    false_count = 0

    for combo in inputs:
        result = f(*combo)
        print(combo, "->", result)
        
        if result:
            true_count += 1
        else:
            false_count += 1
            
    print("True count:", true_count)
    print("False count:", false_count)
```

### 1. Randomly choose constant or balanced

### 3. Return the function that generator creates

```python
def random_constant_balanced():
    """
    This function is responsike for returning either a constant Boolean function or a balanced Boolean fucnction.
    
    The Deutsch-Jozsa algorithm assumes the function is guaranteed to be either:
    
        - Constant
        - Balanced
        
    random_constant_balanced() randomly chooses between those two cases
    
    This function will randomly choose constant or balanced, call the corresponding generator function and return the generated function
    
    Returns:
    function
    
    A function that:
        - takes four Boolean inputs
        - returns a Boolean value
        - is guaranteed to be either constant or balanced
    """
    
    # Randomly choose which type of function to create
    function_type = random.choice(["constant", "balanced"])

    # If the choice is constant, create a constant function
    if function_type == "constant":
        return generate_constant_function()

    # Otherwise, create a balanced function
    else:
        return generate_balanced_function()
```

### Step 1: Generate All Input Combinations

### Before we can test the function f classically, we need to know **all possible inputs**.

### Since f takes four Boolean inputs:

- There are 2⁴ = 16 total input combinations

### How We Generate Inputs

### We can use a simple approach:

1. Consider each input position as a "slot" that can be True or False.
2. Combine all possibilities for the four slots.
3. Store each combination as a tuple of four Boolean values.

### Example: (False, False, False, False)
### (False, False, False, True)
### (True, True, True, True)


### This will give exactly 16 tuples.


```python
# Import product from itertools to generate combinations
from itertools import product

def generate_all_inputs():
    """
    Generate all possible Boolean input combinations for four inputs.

    Returns:
    --------
    list of tuples
        Each tuple contains four Boolean values representing a single input combination.

    Example:
    --------
    [
        (False, False, False, False),
        (False, False, False, True),
        ...
        (True, True, True, True)
    ]
    """
    # Use itertools.product to generate every combination of False and True repeated 4 times
    all_combinations = list(product([False, True], repeat=4))
    
    return all_combinations

# Test the function to ensure it works correctly
inputs = generate_all_inputs()
print("Total input combinations:", len(inputs))
print(inputs)
```

2. `list(product(...))`:

- This allows us to access combinations multiple times

3. The function `generate_all_inputs` returns the **full list of 16 inputs**

### This ensures that later, when we test the function f classically, we will not miss any input.

### Step 3: Implementing the Function `determine_constant_balanced`

Now that we understand how classical testing works, we can turn that idea into a reusable Python function.

The goal of this function is simple:

It will take another function `f` as input and determine whether that function is **constant** or **balanced**.

Recall the definitions:

A **constant function** always returns the same value for every input.

A **balanced function** returns:
- True for exactly half of the inputs
- False for the other half

Since our function takes four Boolean inputs, there are **16 total input combinations**.

Therefore:

- A constant function will return either **16 True values** or **16 False values**
- A balanced function will return **8 True values and 8 False values**

Our strategy will be:

1. Generate all possible inputs
2. Call the function `f` on each input
3. Count the number of True and False outputs
4. Use these counts to determine the type of function

```python
def determine_constant_balanced(f):
    """
    Determine whether a given Boolean function is constant or balanced.

    Parameters:
    
    f : function
        A function that takes four Boolean inputs (a, b, c, d)
        and returns a Boolean value.

    Returns:
    
    str
        Returns the string "constant" if the function always returns
        the same value for all inputs.

        Returns the string "balanced" if the function returns True
        for exactly half of the inputs and False for the other half.

    Explanation:
    
    The function works by evaluating the input function `f` on every
    possible combination of four Boolean inputs. There are 16 possible
    combinations in total.

    By counting how many times the function returns True and False,
    we can determine whether the function is constant or balanced.
    """

    # Generate all 16 possible input combinations
    inputs = generate_all_inputs()

    # Counters to keep track of how many times
    # the function returns True and False
    true_count = 0
    false_count = 0

    # Loop through each possible input combination
    for combo in inputs:

        # Call the function f with the current input values
        # The * operator unpacks the tuple into four arguments
        result = f(*combo)

        # If the function returns True, increase the True counter
        if result:
            true_count += 1

        # Otherwise increase the False counter
        else:
            false_count += 1

    # After testing all inputs we analyze the counts

    # If all outputs are the same, the function is constant
    if true_count == 16 or false_count == 16:
        return "constant"

    # Otherwise, the function must be balanced
    else:
        return "balanced"
```

### Explanation of the Implementation

The function begins by generating all possible input combinations using `generate_all_inputs()`.

This gives us a list of 16 tuples, where each tuple represents a unique combination of four Boolean values.

Next, we create two counters:

- `true_count`
- `false_count`

These counters keep track of how many times the function returns True and False.

We then loop through every input combination and call the function `f`.

The line:

result = f(*combo)

uses Python's unpacking operator `*`.

This converts a tuple like:

(True, False, True, False)

into a function call:

f(True, False, True, False)

Each result is recorded in the appropriate counter.

After all inputs have been tested, we analyze the results.

If every output is the same, the function is constant.

Otherwise, because the problem guarantees only two possibilities, the function must be balanced.

### Why This Method Works

This approach works because we examine the complete behaviour of the function.

By evaluating the function on **all possible inputs**, we eliminate any uncertainty about its behaviour.

This guarantees that our classification of the function as constant or balanced is always correct.

However, this method may not be the most efficient classical approach.

### Efficiency Analysis of the Classical Solution

In the previous section, we implemented a classical algorithm that determines whether a Boolean function is constant or balanced by evaluating the function on every possible input.

This method guarantees that we will always reach the correct conclusion.

However, it raises an important question about **efficiency**.

In computer science, efficiency usually refers to how much work an algorithm must perform in order to solve a problem. One common way to measure this work is by counting how many times a function must be called.

In this case, we are interested in answering the following question:

**What is the maximum number of times we must call the function `f` in order to be absolutely certain whether it is constant or balanced?**

To answer this, we must carefully examine how the function behaves in different situations.

### Number of Possible Inputs

The function `f` takes four Boolean inputs:

f(a, b, c, d)

Each input can take two possible values:

- True
- False

Because there are four inputs, the total number of possible input combinations is:

2⁴ = 16

This means there are exactly **16 different inputs** that could be given to the function.

If we evaluate the function on all 16 inputs, we will completely understand how the function behaves.

This is the approach used in our implementation, and it guarantees a correct result.

### Behaviour of Constant Functions

A constant function always returns the same output regardless of the input.

For example, a constant function might always return:

True

or it might always return:

False

This means that if the function is constant, every time we call the function we will receive the same result.

For example:

f(input₁) → True  
f(input₂) → True  
f(input₃) → True  
f(input₄) → True  

If this pattern continues for all inputs, we know the function is constant.

However, seeing several identical outputs does not immediately prove that the function is constant. It could still be balanced, and we might simply have tested inputs that happen to produce the same result.

### Behaviour of Balanced Functions

A balanced function returns True for exactly half of the inputs and False for the other half.

Since there are 16 possible inputs, this means:

- 8 inputs produce True
- 8 inputs produce False

When testing a balanced function, we may initially observe many identical outputs.

For example:

True  
True  
True  
True  
True  

This does not guarantee that the function is constant, because the remaining inputs may still produce False.

Therefore, the classical algorithm must continue testing until it finds evidence that the outputs are not all the same.

### The Worst Case Scenario

The worst case occurs when the outputs we observe provide the least amount of information about the function.

Imagine that we test the function repeatedly and obtain the following results:

True  
True  
True  
True  
True  
True  
True  
True  

At this point we have tested the function 8 times and every output has been True.

Can we conclude that the function is constant?

No.

The function might still be balanced. It is possible that the remaining 8 inputs produce False.

Because of this uncertainty, we must test the function at least one more time.

### The Key Insight

After observing 8 identical outputs, the function could still be balanced.

However, once we test the function **one more time**, two outcomes are possible.

Case 1: The output is different.

For example:

True  
True  
True

Now we immediately know that the function cannot be constant. Therefore, it must be balanced.

Case 2: The output is the same again.

True  
True  
True  
True  
True

Now we have seen 9 identical outputs.

This means that the function cannot be balanced, because a balanced function can only produce 8 True outputs.

Therefore, the function must be constant.

### Maximum Number of Function Calls

From the reasoning above, we can determine the maximum number of function calls needed to be absolutely certain.

In the worst case, we must evaluate the function **9 times**.

This is because:

- A balanced function can produce at most 8 identical outputs.
- Seeing 9 identical outputs proves the function cannot be balanced.

Therefore, after 9 evaluations we are guaranteed to know whether the function is constant or balanced.

**Maximum classical calls required: 9**

### Connection to Deutsch's Algorithm

This result highlights why Deutsch's algorithm is important.

A classical computer may require up to **9 function evaluations** to be completely certain whether the function is constant or balanced.

In contrast, Deutsch's quantum algorithm can determine the answer using **only one evaluation of the function**.
